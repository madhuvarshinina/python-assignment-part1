```python
# ------------ TASK 1 ----------------

raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

print("\n--- Cleaned Student Profiles ---\n")

for st in raw_students:
    # cleaning name
    name_clean = st["name"].strip().title()

    # converting roll
    rno = int(st["roll"])

    # converting marks string to list
    parts = st["marks_str"].split(",")
    marks = []
    for p in parts:
        marks.append(int(p.strip()))

    # checking if name is valid
    valid = True
    for w in name_clean.split():
        if not w.isalpha():
            valid = False

    if valid:
        print("✓ Valid name")
    else:
        print("✗ Invalid name")

    # printing profile
    print("================================")
    print(f"Student : {name_clean}")
    print(f"Roll No : {rno}")
    print(f"Marks   : {marks}")
    print("================================")

    # special condition
    if rno == 103:
        print("Special Check:")
        print(name_clean.upper())
        print(name_clean.lower())


# ------------- TASK 2 ---------------

student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

print("\n--- Subject-wise Marks ---\n")

for i in range(len(subjects)):
    sub = subjects[i]
    score = marks[i]

    # grade logic (basic)
    if score >= 90:
        grd = "A+"
    elif score >= 80:
        grd = "A"
    elif score >= 70:
        grd = "B"
    elif score >= 60:
        grd = "C"
    else:
        grd = "F"

    print(sub, ":", score, "-", grd)


# totals
tot = sum(marks)
avg = round(tot / len(marks), 2)

print("\nTotal:", tot)
print("Average:", avg)

# highest and lowest
hi = max(marks)
lo = min(marks)

hi_sub = subjects[marks.index(hi)]
lo_sub = subjects[marks.index(lo)]

print("Highest:", hi_sub, "-", hi)
print("Lowest:", lo_sub, "-", lo)


# while loop for adding subjects
print("\n--- Add More Subjects --")

new_count = 0

while True:
    sub_name = input("Enter subject name (or done): ")

    if sub_name.lower() == "done":
        break

    mk_in = input("Enter marks: ")

    if not mk_in.isdigit():
        print("do not crash. Invalid input, try again")
        continue

    mk_val = int(mk_in)

    if mk_val < 0 or mk_val > 100:
        print("do not add invalid entries to the list, Marks should be between 0 and 100")
        continue

    subjects.append(sub_name)
    marks.append(mk_val)
    new_count += 1

print("Added subjects:", new_count)

new_avg = round(sum(marks) / len(marks), 2)
print("Updated Average:", new_avg)


# ---------------- TASK 3 --------------

class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

print("\n--- Class Report ---\n")
print("Name              | Average | Status")
print("----------------------------------------")

p_count = 0
f_count = 0

top_name = ""
top_avg = 0

all_avg = 0

for stu in class_data:
    name = stu[0]
    mks = stu[1]

    avg = round(sum(mks) / len(mks), 2)
    all_avg += avg

    if avg >= 60:
        stat = "Pass"
        p_count += 1
    else:
        stat = "Fail"
        f_count += 1

    print(f"{name:<18} | {avg:6}  | {stat}")

    if avg > top_avg:
        top_avg = avg
        top_name = name

print("\nPassed:", p_count)
print("Failed:", f_count)

print("Topper:", top_name, "-", top_avg)

class_avg = round(all_avg / len(class_data), 2)
print("Class Average:", class_avg)


# ------------ TASK 4 ---------------

essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

print("\n--- String Operations ---\n")

# step 1
clean_essay = essay.strip()
print("Cleaned:", clean_essay)

# step 2
print("\nTitle Case:")
print(clean_essay.title())

# step 3
count = clean_essay.count("python")
print("\nCount of 'python':", count)

# step 4
rep = clean_essay.replace("python", "Python 🐍")
print("\nReplaced:")
print(rep)

# step 5
sent_list = clean_essay.split(". ")
print("\nSentence List:")
print(sent_list)

# step 6
print("\nNumbered Sentences:")
for i in range(len(sent_list)):
    s = sent_list[i]
    if not s.endswith("."):
        s = s + "."
    print(i+1, "-", s)
```
